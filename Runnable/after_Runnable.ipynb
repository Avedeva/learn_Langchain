## After Runnable 


After runnable we chan make many chain and that can work seemlessly as langchain made the under same abstract method


```python
from abc import ABC,abstractmethod

class runnable(ABC):

    @abstractmethod
    def invoke(input_data):
        pass


import random
class fakellm(runnable):
    def __init__(self):
        print("llm created")


    def invoke(self,prompt):
        response_list = [
            "wow","hell no","the fuhhh","faaah","must be the wind"
        ]

        return {'response': random.choice(response_list)}


    def predict(self,prompt):
        response_list = [
            "wow","hell no","the fuhhh","faaah","must be the wind"
        ]

        return {'response': random.choice(response_list)}





class fakeprompttemplate(runnable):
    def __init__(self,template,input_variable):
        self.template = template
        self.input_variable = input_variable


    def invoke(self,input_dict):
        return self.template.format(**input_dict)

    def format(self,input_dict):
        return self.template.format(**input_dict)



class fakestroutparser(runnable):
    def __init__(self):
        print("parser init")

    def invoke(self,result):
        return result['response']

    def parse(self,result):
        return result['response']
```

```python
class runnabelconneter(runnable):
    def __init__(self,runnable_link):

        self.runnable_link = runnable_link


    def invoke(self,input_data):

        for runnable in self.runnable_link:

            input_data = runnable.invoke(input_data)


        return input_data
```

```python
prompt = fakeprompttemplate(
    template = "How my mom would react if i said {confess}",
    input_variable=['confess']
)

llm = fakellm()

parser = fakestroutparser()
```

```python
chain = runnabelconneter([prompt,llm,parser])

chain.invoke({'confess':'I ran 5km today'})
```

```python
class fakeprompttemplate:
    def __init__(self,template,input_variable):
        self.template = template
        self.input_variable = input_variable

    def format(self,input_dict):
        return self.template.format(**input_dict)


# this is a template class does same as the langchain PromptTemplate class
```

```python
template = fakeprompttemplate(
    template= "How my father will react when i say {accomplishment}",
    input_variable=['accomplishment']
)

prompt = template.format({'accomplishment':' I studying 10 hours'})
prompt
```

--- After chain --

```python
class fakellmchain:
    def __init__(self,llm,prompt):
        self.prompt = prompt
        self.llm = llm

    def run(self,input_dict):
        self.prompt = self.prompt.format(input_dict)
        return llm.predict(self.prompt)


# llm chain takes prompt and llm and gives the result langchain did this for many use cases retrieval qa and other uses cases
```

```python
llm = fakellm()

prompt = fakeprompttemplate(
    template = 'How my sister react if said {confessing}',
    input_variable=['confessing']
)

chain = fakellmchain(prompt = prompt,llm = llm)
result = chain.run({'confessing':'I ran 5km in 30 min'})

print("\nPrompt :- ",prompt.format({'confessing':'I ran 5km in 30 min'}))
print("\nReply :- ",result['response'])
```
